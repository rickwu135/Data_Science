```python
#importing libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn import linear_model
from sklearn.linear_model import Ridge, Lasso
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error
from sklearn.preprocessing import StandardScaler
```

```python
#use pandas to read the csv files
train = pd.read_csv('train.csv')
test = pd.read_csv('test.csv')

# some categorical features such as MSZoning, Street
# train.info()
# test.info()

#check out the size of the data
print ("Train data shape:", train.shape)
print ("Test data shape:", test.shape)

#look at a few rows using the DataFrame.head() method
# print(train.head())

#to do some plotting
plt.style.use(style='ggplot')
plt.rcParams['figure.figsize'] = (10, 6)
```

```python
# Data Exploration and Feature Engineering

# Display summary statistics for the 'SalePrice' column in the 'train' dataset
print(train.SalePrice.describe())

# Data is left leaning
plt.hist(train.SalePrice, color='blue')
plt.show()

# Apply a log transformation to the 'SalePrice' data to reduce the skew
target = np.log(train.SalePrice)
plt.hist(target, color='blue')
plt.show()
```

```python
# Working with numerical features

# Select all numeric columns in the 'train' DataFrame and store them
numerical_features = train.select_dtypes(include=[np.number])
print(numerical_features.dtypes)
print("-------------------------------------------------------------------")

# Compute the pairwise correlation of the columns
corr = numerical_features.corr()

# first five features are most correlated with SalePrice, while the next five are least correlated
print(corr['SalePrice'].sort_values(ascending=False)[::1], '\n')
```

```python
high_corr_features=corr.index[abs(corr['SalePrice'])>0.52]
print(f'highly correlated feature:\n',high_corr_features)
print(f'No. of highly correlated features:',len(high_corr_features))
```

```python
# calculate the median of 'SalePrice' for each unique 'OverallQual' value
quality_pivot = train.pivot_table(index='OverallQual', values='SalePrice', aggfunc=np.median)
print(quality_pivot)

# plot showing it
quality_pivot.plot(kind='bar', color='blue')
plt.xlabel('Overall Quality')
plt.ylabel('Median Sale Price')
plt.xticks(rotation=0)
plt.show()
```

```python
# Handling the null values

# sort by number of null values and ranks them in descending order
nulls = pd.DataFrame(train.isnull().sum().sort_values(ascending=False)[:30])
nulls.columns = ['Null Count']
nulls.index.name = 'Feature:'
print(nulls)
```

```python
# display summary statistics for categorical features
print(train.describe(include='object'))
```

```python
# a look at categorical features
# just an example to see as we can see that overall quality is best so this is just practice

#This is about if the houses streets leading to driveway, gravel or paved
print("Original: \n")
print(train.Street.value_counts(), "\n")

#instantiate label envoder
labelencoder = LabelEncoder()

train['enc_street'] = labelencoder.fit_transform(train.Street)
test['enc_street'] = labelencoder.transform(test.Street)

print('Encoded: \n')
print(train.enc_street.value_counts())  # Pave and Grvl values converted into numerical values

# Plot GarageType
type_pivot = train.pivot_table(index='GarageType', values='SalePrice', aggfunc=np.median)
type_pivot.plot(kind='bar', color='blue')
plt.xlabel('GarageType')
plt.ylabel('Sale Price')
plt.show()

# Encode GarageType using LabelEncoder
train['enc_type'] = labelencoder.fit_transform(train['GarageType'])
test['enc_type'] = labelencoder.transform(test['GarageType'])

# Create a mapping dictionary for the encoded values and their corresponding original values
mapping = dict(zip(train['enc_type'], train['GarageType']))

# Plot the encoded GarageType
type_pivot_encoded = train.pivot_table(index='enc_type', values='SalePrice', aggfunc=np.median)
type_pivot_encoded.plot(kind='bar', color='blue')
plt.xlabel('Encoded GarageType')
plt.ylabel('Sale Price')
plt.show()
```

```python
# Dealing with missing values

# Select numeric columns from the train dataset
data = train.select_dtypes(include=[np.number])

# Fill missing values with the mean of the respective column
data.fillna(data.mean(), inplace=True)

# Check if all of columns have 0 null values
print(sum(data.isnull().sum() != 0))

# Check for categorical features as well
print(len(data.select_dtypes(include='object').columns))
```

```python
# separate features
# X = features and y = target

y = np.log(train.SalePrice)
X = data.drop(['SalePrice', 'Id'], axis=1)

# Partition data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
# begin the modelling

# Linear Regression Model
lr = linear_model.LinearRegression()
model = lr.fit(X_train, y_train)
prediction = lr.predict(X_test)

actual_values = y_test
train_score_lr = lr.score(X_train, y_train)
test_score_lr = lr.score(X_test, y_test)
mae_lr = mean_absolute_error(y_test, prediction)
mse_lr = mean_squared_error(y_test, prediction)

print("\nLinear Model\n")
print("The train score for lr model is {}".format(train_score_lr))
print("The test score for lr model is {}".format(test_score_lr))
print("The MAE for lr model is {}".format(mae_lr))
print("The MSE for lr model is {}".format(mse_lr))

# Ridge Regression Model
ridgeReg = Ridge(alpha=10)
model2 = ridgeReg.fit(X_train,y_train)
prediction2 = ridgeReg.predict(X_test)

train_score_ridge = ridgeReg.score(X_train, y_train)
test_score_ridge = ridgeReg.score(X_test, y_test)
mae_ridge = mean_absolute_error(y_test, prediction2)
mse_ridge = mean_squared_error(y_test, prediction2)

print("\nRidge Model\n")
print("The train score for ridge model is {}".format(train_score_ridge))
print("The test score for ridge model is {}".format(test_score_ridge))
print("The MAE for ridge model is {}".format(mae_ridge))
print("The MSE for ridge model is {}".format(mse_ridge))

# Lasso Regression Model
lasso = Lasso(alpha = .0001)
model3 = lasso.fit(X_train,y_train)
prediction3 = lasso.predict(X_test)

train_score_ls = lasso.score(X_train,y_train)
test_score_ls = lasso.score(X_test,y_test)
mae_lasso = mean_absolute_error(y_test, prediction3)
mse_lasso = mean_squared_error(y_test, prediction3)

print("\nLasso Model\n")
print("The train score for ls model is {}".format(train_score_ls))
print("The test score for ls model is {}".format(test_score_ls))
print("The MAE for ls model is {}".format(mae_lasso))
print("The MSE for ls model is {}".format(mse_lasso))
```

```python
# evaluate performance and visualize results
print ("R^2 is:", model.score(X_test, y_test))

#use the model we have built to make predictions on the test data set.
predictions = model.predict(X_test)

#calculates the rmse
print ('RMSE is:', mean_squared_error(y_test, predictions))

#view this relationship between predictions and actual_values graphically with a scatter plot.
actual_values = y_test
plt.scatter(predictions, actual_values, alpha=.75,
            color='b') #alpha helps to show overlapping data
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.title('Linear Regression Model')
plt.show()

print("---------------------------------------------------------------------------------------")

# evaluate performance and visualize results
print ("R^2 is:", model2.score(X_test, y_test))

#use the model we have built to make predictions on the test data set.
predictions = model2.predict(X_test)

#calculates the rmse
print ('RMSE is:', mean_squared_error(y_test, predictions))

#view this relationship between predictions and actual_values graphically with a scatter plot.
actual_values = y_test
plt.scatter(predictions, actual_values, alpha=.75,
            color='b') #alpha helps to show overlapping data
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.title('Ridge Regression Model')
plt.show()

print("---------------------------------------------------------------------------------------")

# evaluate performance and visualize results
print ("R^2 is:", model3.score(X_test, y_test))

#use the model we have built to make predictions on the test data set.
predictions = model3.predict(X_test)

#calculates the rmse
print ('RMSE is:', mean_squared_error(y_test, predictions))

#view this relationship between predictions and actual_values graphically with a scatter plot.
actual_values = y_test
plt.scatter(predictions, actual_values, alpha=.75,
            color='b') #alpha helps to show overlapping data
plt.xlabel('Predicted Price')
plt.ylabel('Actual Price')
plt.title('Lasso Regression Model')
plt.show()
```

```python
# making a submission

# Create a DataFrame for submission
submission = pd.DataFrame()
submission['Id'] = test.Id

# Select the features from the test data for the model as we did above
feats = test.select_dtypes(include=[np.number]).drop(['Id'], axis=1).interpolate()

# Generate predictions for each model
predictions_lr = model.predict(feats)
predictions_ridge = model2.predict(feats)
predictions_lasso = model3.predict(feats)

# Transform the predictions to the correct form
# Apply np.exp() to our predictions because we have taken the logarithm(np.log()) previously
final_predictions_lr = np.exp(predictions_lr)
final_predictions_ridge = np.exp(predictions_ridge)
final_predictions_lasso = np.exp(predictions_lasso)

# Check the difference
print("Linear Regression Model")
print("Original predictions are: \n", predictions_lr[:10], "\n")
print("Final predictions are: \n", final_predictions_lr[:10])

print("\nRidge Regression Model")
print("Original predictions are: \n", predictions_ridge[:10], "\n")
print("Final predictions are: \n", final_predictions_ridge[:10])

print("\nLasso Regression Model")
print("Original predictions are: \n", predictions_lasso[:10], "\n")
print("Final predictions are: \n", final_predictions_lasso[:10])

# Add predictions to the submission DataFrame
submission['SalePrice_LR'] = final_predictions_lr
submission['SalePrice_Ridge'] = final_predictions_ridge
submission['SalePrice_Lasso'] = final_predictions_lasso

# Save the submission DataFrame to a CSV file
# saves all together
submission.to_csv('submission.csv', index=False)
```

```python
# This was to turn it into Kaggle to see quality
# Create a DataFrame for submission (Linear Regression)
submission_lr = pd.DataFrame()
submission_lr['Id'] = test.Id
submission_lr['SalePrice'] = final_predictions_lr
submission_lr.to_csv('submission_lr.csv', index=False)

# Create a DataFrame for submission (Ridge Regression)
submission_ridge = pd.DataFrame()
submission_ridge['Id'] = test.Id
submission_ridge['SalePrice'] = final_predictions_ridge
submission_ridge.to_csv('submission_ridge.csv', index=False)

# Create a DataFrame for submission (Lasso Regression)
submission_lasso = pd.DataFrame()
submission_lasso['Id'] = test.Id
submission_lasso['SalePrice'] = final_predictions_lasso
submission_lasso.to_csv('submission_lasso.csv', index=False)
```
